```python
from gensim import corpora
data = df_new

text_column = data['Tweets']
nlp = spacy.load("en_core_web_sm")

tokenized_texts = []
for text in text_column:
    doc = nlp(text)
    tokens = [token.text.lower() for token in doc if token.is_alpha and not token.is_stop]
    tokenized_texts.append(tokens)

dictionary = corpora.Dictionary(tokenized_texts)

corpus = [dictionary.doc2bow(tokens) for tokens in tokenized_texts]
```

```python
from gensim.models import LdaModel

num_topics = 5
lda_model = LdaModel(corpus, num_topics=num_topics,id2word=dictionary, passes=10, alpha='auto', eta='auto')

for topic_id in range(num_topics):
    print(f"Topic {topic_id}: {lda_model.print_topic(topic_id)}")
```

```python
row_index = 53
new_document = data.loc[row_index, 'Tweets']

doc = nlp(new_document)
tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
new_bow = dictionary.doc2bow(tokens)

topics = lda_model.get_document_topics(new_bow)

for topic in topics:
    topic_id, prob = topic
    print(f"Topic {topic_id}: Probability: {prob:.4f}")
```

```python
import matplotlib.pyplot as plt

row_index = 53
new_document = data.loc[row_index, 'Tweets']

doc = nlp(new_document)
tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
new_bow = dictionary.doc2bow(tokens)

topics = lda_model.get_document_topics(new_bow)

topic_probs = [prob for _, prob in topics]

topic_labels = [f"Topic {i}" for i in range(len(topic_probs))]

plt.figure(figsize=(10, 6))
plt.bar(topic_labels, topic_probs, color='skyblue')
plt.xlabel('Topics')
plt.ylabel('Probability')
plt.title('Topic Distribution for New Document')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
num_topics = 5

for topic_id in range(num_topics):
    top_keywords = lda_model.print_topic(topic_id)
    top_words = [word.split('*')[1].strip('" ') for word in top_keywords.split('+')]

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(top_words))
    plt.figure(figsize=(8, 4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f"Word Cloud for Topic {topic_id}")
    plt.axis('off')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
file_path = '/content/drive/MyDrive/new.csv'
data = pd.read_csv(file_path)

topic_distribution = defaultdict(list)
for idx, row in data.iterrows():
    document = row['Tweets']
    doc_tokens = [token.lemma_ for token in nlp(document) if token.is_alpha and not token.is_stop]
    doc_bow = dictionary.doc2bow(doc_tokens)
    doc_topics = lda_model.get_document_topics(doc_bow)
    for topic_id, prob in doc_topics:
        topic_distribution[f"Topic {topic_id}"].append(prob)

avg_topic_probs = {topic: sum(probs) / len(probs) for topic, probs in topic_distribution.items()}

sorted_topics = sorted(avg_topic_probs.items(), key=lambda x: x[1], reverse=True)
topic_labels, avg_probs = zip(*sorted_topics)

plt.figure(figsize=(12, 6))
plt.bar(topic_labels, avg_probs, color='skyblue')
plt.xlabel('Topics')
plt.ylabel('Average Probability')
plt.title('Average Topic Distribution Across Documents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
